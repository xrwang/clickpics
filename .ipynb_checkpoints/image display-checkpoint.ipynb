```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import os
import glob
from PIL import Image 
```

```python
imgstring = 'training/IMG_5975.JPG'
img = mpimg.imread(imgstring) 
plt.figure(figsize=(8.0, 10.0))
plt.axis('off')
plt.imshow(img) 
plt.text(0, 0,imgstring , fontsize=12)
plt.savefig(imgstring+'a.png')
```

```python
# Read Images recursively
  
for filepath in glob.iglob('women/*.PNG'):
    img = mpimg.imread(filepath) 
    plt.axis('off')
    plt.imshow(img) 
    plt.text(0, 0,imgstring , fontsize=12)
    plt.savefig('books_read.png')


```
