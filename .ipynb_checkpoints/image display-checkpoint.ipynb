```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import os
import glob
import psutil
from PIL import Image 
import csv
import pandas as pd 
```

```python
df = pd.read_csv('judgements-final.csv')
```

```python
def img_url_chopper(row):
    url = row['image_url']
    split = url.split('/')
    img_name = split[4]
    return img_name
```

```python
df['image_name'] = df.apply (lambda row: img_url_chopper(row), axis=1)
df
```

```python
imgstring = 'training/IMG_5975.JPG'
img = mpimg.imread(imgstring) 
img_name = imgstring.split('/')[1]
rows = df.loc[df['image_name'] == img_name]
avatar = rows['does_this_image_contain_an_avatar']
country = rows['_country']
ip = rows['_ip']
plt.figure(figsize=(8, 22.0))
plt.axis('off')
plt.imshow(img) 
plt.text(0, -250, avatar, fontsize=12)
plt.text(0, -60, country, fontsize =12)
plt.text(0, 0,imgstring , fontsize=12)

plt.savefig(imgstring+'a.png')
```

```python
# Read Images recursively
  
for filepath in glob.iglob('training/*.JPG'):
    
    img = mpimg.imread(filepath) 
    
    img_name = filepath.split('/')[1]
    rows = df.loc[df['image_name'] == img_name]
    avatar = rows['does_this_image_contain_an_avatar']
    country = rows['_country']
    ip = rows['_ip']
    plt.figure(figsize=(8, 22.0))
    plt.axis('off')
    plt.text(0, -250, avatar, fontsize=12)
    plt.text(0, -60, country, fontsize =12)
    plt.text(0, 0,imgstring , fontsize=12)
    #plt.figure(figsize=(8.0, 10.0))
    #plt.axis('off')
    #plt.text(0, 0,filepath, fontsize=12)
    plt.imshow(img) 
    plt.savefig(filepath+'a.png')
    plt.clf()

```
